```python
def int_operation(int1, int2):
    if int1 * int2 <= 1000:
        return int1 * int2
    else:
        return int1 + int2
```

```python
# This block should not return any error
assert int_operation(0,0) == 0
assert int_operation(10, 10) == 100
assert int_operation(0, 100000000000) == 0
assert int_operation(1, 100000000000) == 100000000001
assert int_operation(40, 30) == 70
assert int_operation(10, 100) == 1000
assert int_operation(10, 101) == 111
```

```python
def remove_chars(string_input, n):
    return string_input[n:]
```

```python
assert remove_chars("pynative", 4) == "tive"
assert remove_chars("pynative", 2) == "native"
assert remove_chars("pynative", 0) == "pynative"
assert remove_chars("pynative", 100) == ""
```

```python
def is_first_last_same(x):
    return x[0] == x[-1]
```

```python
assert is_first_last_same([10])
assert not is_first_last_same([1,2,3,4])
assert is_first_last_same([1,2,3,1])
```

Exercise 5: Display numbers divisible by 5 from a list

```python
def display_number_divisible(x):
    for num in x:
        if num % 5 == 0:
            print(num)
```

```python
display_number_divisible([10, 20, 33, 46, 55])
```

```python
def is_palindrome_number(x):
    str_num = str(x)
    str_num_reverse = str_num[::-1]
    return str_num == str_num_reverse
```

```python
assert is_palindrome_number(121)
assert not is_palindrome_number(125)
assert is_palindrome_number(1)
```

```python
def calculate_income_tax(x):
    if x < 10000:
        return 0
    elif x < 20000:
        return (x - 10000) * 0.1
    else:
        return 10000 * 0.1 + (x - 20000) * 0.2
```

```python
assert calculate_income_tax(500) == 0
assert calculate_income_tax(10001) == 0.1
assert calculate_income_tax(45000) == 6000
```

```python
def exponent(base, power):
    return base ** power
```

```python
assert exponent(2,5) == 32
assert exponent(5,4) == 625
```
